```python
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python

from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
```

```python
def train_test(test_ratio):
  (train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
  train_images = train_images/255
  test_images = test_images/255

  data_labels = np.concatenate((train_labels, test_labels))
  data_images = np.concatenate((train_images, test_images))

  train_images, test_images, train_labels, test_labels = train_test_split(data_images, data_labels, test_size = test_ratio)
  return train_images, test_images, train_labels, test_labels
```

```python
train_images, test_images, train_labels, test_labels = train_test(0.2)
```

```python
test_labels.size
```

```python
categories = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
    plt.xlabel(categories[train_labels[i][0]])
plt.show()

```

```python
def create_model(filtras, aktyvacijos_funk, dense_activ):
  model = models.Sequential()
  model.add(layers.Conv2D(32, filtras, activation = aktyvacijos_funk, input_shape = (32, 32, 3)))
  model.add(layers.MaxPooling2D(2, 2))
  model.add(layers.Conv2D(64, filtras, activation = aktyvacijos_funk, input_shape = (32, 32, 3)))
  model.add(layers.MaxPooling2D(2, 2))
  
  model.add(layers.Flatten())
  model.add(layers.Dense(64, activation = dense_activ))
  model.add(layers.Dense(10))

  return model

```

```python
model = create_model((3, 3), "relu", "relu")
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
              loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics = 'accuracy')
```

# New Section

```python
model.fit(train_images, train_labels, epochs = 10, validation_data=(test_images, test_labels), batch_size=24)
```

```python
model.predict(test_images)
```

```python
plt.plot(model.history['accuracy'], label='accuracy')
plt.plot(model.history'val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
```
